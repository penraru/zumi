```python
from picamera import PiCamera
import cv2
import IPython.display 
import PIL.Image

# Take a photo with the camera, and save it
camera = PiCamera()
camera.capture("hi.jpg")
camera.close()

# Load the image and display it here
image = cv2.imread("hi.jpg")
cv2.ellipse(image,(400,256),(100,50),0,0,180,255,-1)
image = cv2.circle(image,(350,179), 12, (255,0,0), -1)
image = cv2.circle(image,(440,179), 12, (255,0,0), -1)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
IPython.display.display(PIL.Image.fromarray(image))
```
